```
from pyspark.sql.functions import regexp_replace, regexp_extract, col

regex_string = "^[A-Z][A-Z]$"
```

```
lc_df = lc_df.withColumn("addr_state_cleaned",regexp_extract(col("addr_state"), regex_string, 0))
```

```
display(lc_df.groupBy('addr_state_cleaned').count())
```

```
from pyspark.sql.functions import when, isnan, count, col, log

lc_df = lc_df.withColumn('exposure', when(lc_df.bad_loan == 'No', col('revol_util_cleaned')).otherwise(-10*col('revol_util_cleaned')))

display(lc_df)
                         
```

```
display(lc_df.groupBy('bad_loan','grade').agg({"exposure":'sum'}))
```

```
from pyspark.sql.functions import regexp_replace, regexp_extract, col
lc_df = lc_df.withColumn("dti_cleaned", regexp_extract(col('dti_cleaned'), "\\d+",0))
```

```
# this visualzations are from databricks as it has a zeppelin kinda notebooks, but on spark you can use pandas viz libraries to visualize the data
pd_df = lc_df.toPandas()
#toPandas is very costly as it gets all the data from distributed executors to driver memory. 
#what pandas does is it pickles the data and sends to driver and unpickles it there. The performance is very bad

#Spark 2.X there is arrow execution which is very faster.
```

```
pd_df.dtypes
```

```
import matplotlib.pyplot as plt
import seaborn as sns
plt.clf() #clear current figure
sns.distplot(pd_df.loc[pd_df.dti_cleaned.notnull(), 'dti_cleaned'])

plt.xlabel('dti')
plt.ylabel('count')
display()
```

```
# Enable Arrow-based columnar data transfers
spark.conf.set("spark.sql.execution.arrow.enabled", "true")
```
