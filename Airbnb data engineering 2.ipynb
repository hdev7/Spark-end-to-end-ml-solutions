```
#creating dataframes from the tables
df_listing = spark.table('listing')
df_nhood = spark.table('df_nhood')
df_review = spark.table('df_review')
```

```
#Price contains $ hence it is being treated as string. we should correct it.

from pyspark.sql.types import FloatType

def trim_char(string):
  return string.strip('$')

#registering udf to be used in spark sql
spark.udf.register('trim_func',trim_char)

#registering udf to be used in spark dataframes
trim_func_udf = udf(trim_char)
```

```
df_listing.select('property_type',trim_func_udf('price').cast(FloatType()).alias('price_f')).groupBy('property_type').agg(({'price_f':'average'})).show()
```

```
display(df_listing.select('property_type',trim_func_udf('price').cast(FloatType()).alias('price_f')).groupBy('property_type').agg(({'price_f':'average'})))
```

#### Percentage of Listing in Neighbourhoods

```
import pyspark.sql.functions as f
from pyspark.sql.window import Window

df_neigh_count = df_listing.groupBy('neighbourhood').agg(({'neighbourhood':'count'})).withColumnRenamed("count(neighbourhood)","count")

df_neigh_count = df_neigh_count.withColumn('percentage', 
                                          round(f.col('count')/f.sum('count').over(Window.partitionBy())*100,3))

display(df_neigh_count)
```

```
display(df_neigh_count.orderBy('percentage',ascending = False))
```

#### Join neighbourhood to get neigh_group and plot

```
nei_group = df_neigh_count.alias("nc").join(df_nhood.alias('ne'), col('nc.neighbourhood') == col('ne.neighbourhood')).select('nc.*','ne.neighbourhood_group')

display(nei_group)
```

```
#sort will do in individual exceutors
#orderby will get the results from all the executors and then sort the data

display(nei_group.sort(desc('percentage')))
```

```
#filtering the rows which have proper reviewer_id

df_review_fil = df_review.filter(col('reviewer_id').cast('int').isNotNull())

display(df_review_fil)
```

```
df_review_fil.printSchema()
```

```
df_review_cnt = df_review_fil.groupBy('reviewer_id').count().filter(col('count') > 30)

display(df_review_cnt)
```

```
from datetime import datetime
from pyspark.sql.types import DateType

dfunc = udf (lambda x: datetime.strptime(x,'%Y-%m-%d'), DateType())

spark.udf.register('datec',dfunc)
```

```
df_review_fil.withColumn('datetime', dfunc(col('date'))).show()
```

```
df_review_dt = df_review_fil.withColumn('datetime', dfunc(col('date'))).groupBy('reviewer_id').agg(min('datetime'),max('datetime'))

display(df_review_dt)
```

```
df_review_summary = df_review_cnt.join(df_review_dt,"reviewer_id").withColumn('date_diff',datediff('max(datetime)','min(datetime)')).sort(desc('count'))
```

```
display(df_review_summary)
```

```
df_review_summary.createOrReplaceTempView('review_summary')
```

#### Rank the top 2 listing in each neighbourhood based on reviews

```
%sql

select distinct listing_id,neighbourhood, price,count(listing_id) over(partition by listing_id) as review_cnt from (
select listing_id, neighbourhood, price, reviewer_id from listing a left join df_review b on a.id = b.listing_id) order by review_cnt desc
```

```
%sql 

select listing_id,neighbourhood,price,review_cnt,
dense_rank() over (partition by neighbourhood order by review_cnt desc) as rank
from 
(
select distinct listing_id,neighbourhood, price,count(listing_id) over(partition by listing_id) as review_cnt from 
(
select listing_id, neighbourhood, price, reviewer_id from listing a left join df_review b on a.id = b.listing_id
)
)
```

```
%sql 

select listing_id,neighbourhood,price,review_cnt,rank
from (
select listing_id,neighbourhood,price,review_cnt,
dense_rank() over (partition by neighbourhood order by review_cnt desc) as rank
from 
(
select distinct listing_id,neighbourhood, price,count(listing_id) over(partition by listing_id) as review_cnt from 
(
select listing_id, neighbourhood, price, reviewer_id from listing a left join df_review b on a.id = b.listing_id
)
)) where rank<=2
```

#### Let us see some analytical functions in Spark

```
%sql

select reviewer_id,curr_date,prev_date from (
select reviewer_id, datec(date) as curr_date, lag(datec(date)) over(partition by reviewer_id order by datec(date)) as prev_date 
from df_review where date is not null and date like '%2017%' order by datec(date)
) where prev_date is not null
```

```
%sql

-- To understand the percentile of the spending customers
select reviewer_id,total_spent, CUME_DIST() over(order by total_spent) as cum_dist from
(select reviewer_id, sum(CAST(trim_func(l.price) as float)) as total_spent from df_review as r join listing as l on r.reviewer_id = l.id group by reviewer_id order by 2 desc limit 20) where total_spent is not null
```
